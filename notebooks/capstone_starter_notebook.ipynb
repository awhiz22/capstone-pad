## 6. Data Cleaning

Use this section to clean your dataset.

Common cleaning tasks:
- Fix column names
- Convert data types
- Handle missing values
- Remove duplicates
- Fix inconsistent categories
- Create helpful new columns

Document every major cleaning decision in markdown.


````python
# Make a copy before cleaning

df_clean = ames.copy()
````

````python
# Optional: standardize column names

df_clean.columns = df_clean.columns.str.strip()

print(df_clean.columns.tolist())
````

````python
# TODO: Convert date columns if your dataset has them
# Example:
# df_clean["date"] = pd.to_datetime(df_clean["date"])

df_clean.columns = df_clean.columns.str.strip()

print("Column names cleaned!")
print(df_clean.columns.tolist())


````

````python
# TODO: Handle missing values
# Example strategies:
# df_clean["numeric_column"] = df_clean["numeric_column"].fillna(df_clean["numeric_column"].median())
# df_clean["category_column"] = df_clean["category_column"].fillna("Unknown")

print("Missing values before cleaning:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])

numeric_cols = df_clean.select_dtypes(include='number').columns
for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].median())

categorical_cols = df_clean.select_dtypes(exclude='number').columns
for col in categorical_cols:
    df_clean[col] = df_clean[col].fillna('Unknown')

print("\nMissing values after cleaning:")
print(df_clean.isnull().sum().sum())

````

````python
# TODO: Remove duplicate rows if appropriate
# df_clean = df_clean.drop_duplicates()

print("Duplicate rows before:", df_clean.duplicated().sum())

df_clean = df_clean.drop_duplicates()

print("Duplicate rows after:", df_clean.duplicated().sum())
print("Shape after removing duplicates:", df_clean.shape)
````

````python
# TODO: Check categorical columns for inconsistent labels
# Example:
# df_clean["category_column"].value_counts()
print("Sale_Condition values:")
print(df_clean["Sale_Condition"].value_counts())

print("\nSale_Type values:")
print(df_clean["Sale_Type"].value_counts())

print("\nCentral_Air values:")
print(df_clean["Central_Air"].value_counts())

print("\nOverall_Cond values:")
print(df_clean["Overall_Cond"].value_counts())
# Replace "category_column" with an actual column name
print("Neighborhood values:")
print(df_clean["Neighborhood"].value_counts())

print("\nHouse_Style values:")
print(df_clean["House_Style"].value_counts())

print("\nGarage_Type values:")
print(df_clean["Garage_Type"].value_counts())
````

````python
# Optional: create new useful columns
# Example for e-commerce dataset:
# df_clean["estimated_revenue"] = (
#     df_clean["product_price"] 
#     * df_clean["quantity"] 
#     * (1 - df_clean["discount_applied"])
# )
df_clean["house_age"] = df_clean["Year_Sold"] - df_clean["Year_Built"]

df_clean["total_baths"] = (df_clean["Full_Bath"] + 
                           df_clean["Half_Bath"] + 
                           df_clean["Bsmt_Full_Bath"] + 
                           df_clean["Bsmt_Half_Bath"])

print("New columns added!")
print(df_clean[["Year_Built", "Year_Sold", "house_age", "total_baths"]].head())

````

### 7.2 Group-Based Analysis

Use `.groupby()` to compare groups.

Examples:
- Average revenue by customer segment
- Return rate by product category
- Average price by neighborhood
- Life expectancy by continent


````python
avg_by_neighborhood = df_clean.groupby("Neighborhood")["Sale_Price"].mean().sort_values(ascending=False).head(10)

avg_by_neighborhood.plot(
    kind="bar",
    figsize=(12, 5),
    title="Top 10 Neighborhoods by Average Sale Price",
    color="steelblue"
)
plt.xlabel("Neighborhood")
plt.ylabel("Average Sale Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
avg_by_central_air = df_clean.groupby("Central_Air")["Sale_Price"].mean().sort_values(ascending=False)

avg_by_central_air.plot(
    kind="bar",
    figsize=(8, 5),
    title="Average Sale Price by Central Air",
    color="steelblue"
)
plt.xlabel("Central Air")
plt.ylabel("Average Sale Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

## Final Project Summary

The goal of this project was to figure out what actually drives home sale prices in Ames Iowa and how buyers and 
sellers can use that information to make smarter decisions. The dataset had 2930 rows and 75 columns covering a wide range of home features. Before diving into analysis the data was cleaned by handling missing values, removing 
duplicates and stripping column names of extra whitespace.

The analysis looked at sale price distribution, neighborhood comparisons, the impact of central air and the relationship between living area and price. The strongest finding was that above ground living area has a 0.71 correlation with sale price making it the most reliable predictor. Central air was also a big one — homes with it sell for almost double compared to homes without it. Northridge, Stone_Brook and Northridge_Heights came out as the priciest neighborhoods.

Based on the findings sellers without central air should seriously consider adding it before listing. Buyers looking for value should look at larger homes in more affordable neighborhoods rather than chasing premium zip codes.

That said this analysis has limits. Correlation does not mean causation and the dataset only covers Ames Iowa so 
results may not apply elsewhere. Simple imputation was used for missing values which could affect accuracy. Next steps would be building a regression model and digging deeper into home condition and feature interactions.
